```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
lr=pd.read_csv('/content/Salary_dataset.csv')
lr.info()
```

```python
lr.drop(columns='Unnamed: 0',axis=1,inplace=True)
```

```python
#lr.nunique()
#lr
print(type(lr['Salary']))
```

```python
x=lr['YearsExperience'].values.reshape(-1,1)
#y=lr['Salary'].values.reshape(-1,1)
y=lr['Salary'].squeeze() # Convert y to 1D array, # OR use: y = y.values.ravel()
#y = y.squeeze()
print(x,y)
```

```python
clf = LinearRegression()
clf.fit(x,y)
print(clf.score(x,y)*100)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='YearsExperience', y='Salary', data=lr)
plt.plot(x,clf.predict(x),color='green')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='YearsExperience', y='Salary', data=lr)
plt.title('Salary vs. Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
